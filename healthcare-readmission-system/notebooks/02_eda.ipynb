```python
import sys
from pathlib import Path
from IPython.display import Image, display, Markdown

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT / "backend"))

from app.analytics.eda import load_eda_data, run_full_eda

df = load_eda_data(ROOT / "data" / "processed" / "cleaned_data.csv")
print(f"Rows: {len(df):,} | 30-day readmit rate: {df['readmitted_30_days'].mean()*100:.2f}%")
df.head(3)
```

```python
report = run_full_eda(plots_dir=ROOT / "reports" / "eda_plots")
PLOTS = ROOT / "reports" / "eda_plots"

plot_files = sorted(PLOTS.glob("*.png"))
for plot_path, item in zip(plot_files, report["insights"]):
    display(Markdown(f"### {item['chart']}"))
    display(Image(filename=str(plot_path), width=750))
    display(Markdown(f"**Business Insight:** {item['insight']}"))
    display(Markdown("---"))
```
